Необходимо сгенерировать случайную матрицу n > 100, провести сингулярное разложение, выполнить апроксимацию матрицы с рангом r меняя значение в интервале[2, r]. Для каждого r вычислить ошибку

```python
import numpy as np
import matplotlib.pyplot as plt

n = 125

np.random.seed(42)
A = np.random.rand(n, n)

U, s, Vt = np.linalg.svd(A, full_matrices=False)

errors = []

for r in range(2, n+1):
    U_r = U[:, :r]
    s_r = s[:r]
    Vt_r = Vt[:r, :]

    A_approx = U_r @ np.diag(s_r) @ Vt_r
    errors.append(np.sum((A - A_approx) ** 2) ** 0.5)

plt.plot(errors)

```

## Посчитайте градиент функции и найдите критические точки

$f(x) = x_1^3-2*x_1*x_2+x_2^2-3*x_1-2*x_2, x \in R^2$

градиент
* $\frac{\partial f}{\partial x_1} = 3*x_1^2 - 2*x_2 -3$
* $\frac{\partial f}{\partial x_2} = -2*x_1 + 2 * x_2 - 2$

### Проверте что функция удовлетворяет уравнению

Функция: $f(x) = ln(\sqrt(x_1)+\sqrt(x_2))$

Выражение: $x_1 * \frac{\partial f}{\partial x_1} + x_2 * \frac{\partial f}{\partial x_2} = \frac{1}{2}$ 

Найдем частные производные заданной функции
* $\frac{\partial f}{x_1} = \frac{1}{2 * (\sqrt(x_1) + \sqrt(x_2)) * \sqrt(x_1)}$ 
* $\frac{\partial f}{x_2} = \frac{1}{2 * (\sqrt(x_1) + \sqrt(x_2)) * \sqrt(x_2)}$

Подставим частные производные в выражение
$$
x_1 * \frac{\partial f}{\partial x_1} + x_2 * \frac{\partial f}{\partial x_2} = x_1 * \frac{1}{2 * (\sqrt(x_1) + \sqrt(x_2)) * \sqrt(x_1)} + x_2 * \frac{1}{2 * (\sqrt(x_1) + \sqrt(x_2)) * \sqrt(x_2)} = \frac{1}{2 * (\sqrt(x_1) + \sqrt(x_2))}*(\frac{x_1}{\sqrt(x_1)} + \frac{x_2}{\sqrt{x_2}}) = \frac{1}{2 * (\sqrt(x_1) + \sqrt(x_2))} * (\sqrt(x_1) + \sqrt(x_2)) = \frac{1}{2}
$$

### Найти первый дифференциал и градиент куба евклидовой нормы

$f(x) = \frac{1}{3} \|x\|_2^3 = \frac{1}{3} (\sum(x_i^2))^\frac{3}{2}$

найдем частную производную по $x_i$

$\frac{\partial f}{\partial x_i} = \frac{1}{3} * \frac{3}{2} * (\sum(x_i^2))^\frac{1}{2} * 2 * x_i = \|x\| * x_i $

градиент функции

$\nabla f = (\|x\| * x_1, \|x\| * x_2, ..., \|x\| * x_n) = \|x\| x$

аналогично первый дифференциал

$df = \sum{\frac{\partial f}{\partial x_i}*dx_i} = \sum{\|x\| x_i * dx_i} = \|x\| x dx$

### Найти первый дифференциал и градиент евклидовой нормы

$f(x) = \|x\|_2 = (\sum(x_i^2))^\frac{1}{2}$

найдем частную производную по $x_i$

$\frac{\partial f}{\partial x_i} = \frac{1}{2 * (\sum(x_i^2))^\frac{1}{2}} * 2 * x_i = \frac{x_i}{\|x\|} $

градиент функции

$\nabla f = (\frac{x_1}{\|x\|}, \frac{x_2}{\|x\|}, ..., \frac{x_n}{\|x\|}) = \frac{x}{\|x\|}$

аналогично первый дифференциал

$df = \sum{\frac{\partial f}{\partial x_i}*dx_i} = \sum{\frac{x_i * dx_i}{\|x\|}} = \frac{x dx}{\|x\|}$

### Найти первый дифференциал и градиент функции

Функция: $f(x) = \|Ax\|_2,  A \in R^{mxn}$

рассмотрим выражение $Ax$ как отдельную функцию $y(x)$, тогда для нахождения производной можно воспользоваться правилом $f(y(x))'=f'(y)*y'(x)$

$\frac{\partial f}{\partial x} = \frac{\partial f}{\partial y}*\frac{\partial y}{\partial x}$

производная евклидовой нормы была найдена ранее $\frac{\partial f}{\partial y} = \frac{y}{\|y\|} = \frac{A*x}{\|A*x\|}$

производная линейного преобразования $A*x$ равно самой матрица $A$, соответственно что градиент данной функции равен:

$\nabla f  = \frac{A*x}{\|A*x\|}*A$

аналогично первый дифференциал:

$df = \sum{\frac{\partial f}{\partial x_i}*dx_i} = \frac{A*x}{\|A*x\|}*A*dx$


### Найти первый дифференциал и градиент функции

Функция: $f(x) = -e^{-x^Tx},  x \in R^n$

рассмотрим выражение $-x^Tx$ как отдельную функцию $y(x)$, тогда для нахождения производной можно воспользоваться правилом $f(y(x))'=f'(y)*y'(x)$

$\frac{\partial f}{\partial x} = \frac{\partial f}{\partial y}*\frac{\partial y}{\partial x}$

производная экспоненты $\frac{\partial f}{\partial y} = \frac{\partial (-e^y)}{\partial y} = -e^y = -e^{-x^Tx}$

производная скалярного произведения векторов $\frac{\partial y}{\partial x} = \frac{\partial -x^Tx}{\partial x} = -2x$

соответственно что градиент данной функции равен:

$\nabla f  = -e^{-x^Tx} (-2x) = 2e^{-x^Tx}x$

аналогично первый дифференциал:

$df = 2e^{-x^Tx}xdx$
