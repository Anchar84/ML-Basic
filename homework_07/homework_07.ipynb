# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr1 = np.zeros(10)
arr1[4] = 1
print(f"arr1 = {arr1}\n")

arr12 = arr1.reshape(2, 5)
print(f"arr12 = {arr12}\n")
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr2 = np.arange(10, 50)
print(arr2)
arr21 = arr2[::-1]
print(arr21)
arr22 = arr2[arr2 % 2 == 0]
print(arr22)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr3 = np.arange(9).reshape(3, 3)
print(f"arr3 = {arr3}\n")
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr4 = np.random.randint(0, 100, size=(4, 3, 2))
print(f"arr4 = {arr4}\n")
print(f"min = {arr4.min()}")
print(f"max = {arr4.max()}")
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr5 = np.random.randint(0, 100, size=(6, 4))
arr51 = np.random.randint(0, 100, size=(4, 3))

print(f"arr5 = {arr5}\n")
print(f"arr51 = {arr51}\n")

print(f"arr5 * arr 51 = {arr5 @ arr51}\n")
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr6 = np.random.random((7, 7))
print(f"arr6 = {arr6}\n")

mean = arr6.mean()
std = arr6.std()
print(f"mean = {mean}")
print(f"std = {std}")

arr61 = (arr6 - mean) / std
print(f"arr61 = {arr61}\n")
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
import requests
```

### Загрузите датасет Tips из набора датасетов seaborn

```python

# Почему то sns.load_dataset('tips') выдавал ошибку (не совсем понятно - проверка ssl или timeout)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

# Загрузка с отключенной SSL проверкой
response = requests.get(url)
tips = pd.read_csv(io.StringIO(response.text))
```

```python
print(f"max total_bil = {np.array(tips["total_bill"]).max()}")
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
total_bill_mean = np.array(tips["total_bill"]).mean()
print(f"total_bill mean = {total_bill_mean}")
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
total_bill_median = np.median(tips["total_bill"])

tips_above_median = tips[tips["total_bill"] > total_bill_mean]
male_total_bill_mean = np.array(tips_above_median[tips_above_median["sex"] == "Male"]["total_bill"]).mean()
female_total_bill_mean = np.array(tips_above_median[tips_above_median["sex"] == "Female"]["total_bill"]).mean()

print(f"male_total_bill_mean = {male_total_bill_mean}")
print(f"female_total_bill_mean = {female_total_bill_mean}")
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips["smoker"] = (tips["smoker"] == "Yes").astype(int)
print(tips.head())
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.hist(tips["total_bill"])
plt.title("Распределение признака total_bill")
plt.xlabel("total_bill")
plt.ylabel("частота")

plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.figure(figsize=(10, 6))
plt.scatter(tips["total_bill"], tips["tip"])
plt.title('Зависимость tip от total_bill')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.grid(True, alpha=0.3)
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(tips, hue ='day')
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
plt.figure(figsize=(10, 6))
plt.scatter(tips["total_bill"], tips["day"])
plt.title('Зависимость total_bill от day')
plt.xlabel('total_bill')
plt.ylabel('day')
plt.grid(True, alpha=0.3)
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
lunch_tips = tips[tips['time'] == 'Lunch']['tip']
dinner_tips = tips[tips['time'] == 'Dinner']['tip']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('Распределение чаевых (tip) по времени (time)', fontsize=16, fontweight='bold')

ax1.hist(lunch_tips, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
ax1.set_title('Lunch (Обед)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Размер чаевых ($)', fontsize=12)
ax1.set_ylabel('Частота', fontsize=12)
ax1.grid(True, alpha=0.3)
ax1.text(0.05, 0.95, f'n = {len(lunch_tips)}', transform=ax1.transAxes, 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8),
         fontsize=12, verticalalignment='top')

ax2.hist(dinner_tips, bins=15, color='lightcoral', edgecolor='black', alpha=0.7)
ax2.set_title('Dinner (Ужин)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Размер чаевых ($)', fontsize=12)
ax2.set_ylabel('Частота', fontsize=12)
ax2.grid(True, alpha=0.3)
ax2.text(0.05, 0.95, f'n = {len(dinner_tips)}', transform=ax2.transAxes,
         bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8),
         fontsize=12, verticalalignment='top')

max_tip = tips['tip'].max()
ax1.set_xlim(0, max_tip + 1)
ax2.set_xlim(0, max_tip + 1)

plt.tight_layout()
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Взаимосвязь между Total Bill и Tip по полу (раскраска по Smoker)', fontsize=16, fontweight='bold')

male_data = tips[tips['sex'] == 'Male']
female_data = tips[tips['sex'] == 'Female']

print(f"male_data['smoker'].unique() = {tips['smoker'].unique()}")

for smoker in male_data['smoker'].unique():
    subset = male_data[male_data['smoker'] == smoker]
    color = 'red' if smoker else 'blue'
    label = 'Smoker' if smoker else 'Non-Smoker'
    
    ax1.scatter(subset['total_bill'], subset['tip'], 
               c=color, alpha=0.7, s=60, label=label, edgecolors='black', linewidth=0.5)

ax1.set_title('Male (Мужчины)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Total Bill (Общий счет, $)', fontsize=12)
ax1.set_ylabel('Tip (Чаевые, $)', fontsize=12)
ax1.legend()
ax1.grid(True, alpha=0.3)

z_male = np.polyfit(male_data['total_bill'], male_data['tip'], 1)
p_male = np.poly1d(z_male)
ax1.plot(male_data['total_bill'], p_male(male_data['total_bill']), "k--", 
         linewidth=2, alpha=0.8, label='Линия регрессии')

corr_male = male_data['total_bill'].corr(male_data['tip'])
ax1.text(0.05, 0.95, f'Корреляция: {corr_male:.3f}\nn = {len(male_data)}', 
         transform=ax1.transAxes, 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8),
         verticalalignment='top')

for smoker in female_data['smoker'].unique():
    subset = female_data[female_data['smoker'] == smoker]
    color = 'red' if smoker else 'blue'
    label = 'Smoker' if smoker else 'Non-Smoker'
    
    ax2.scatter(subset['total_bill'], subset['tip'], 
               c=color, alpha=0.7, s=60, label=label, edgecolors='black', linewidth=0.5)

ax2.set_title('Female (Женщины)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Total Bill (Общий счет, $)', fontsize=12)
ax2.set_ylabel('Tip (Чаевые, $)', fontsize=12)
ax2.legend()
ax2.grid(True, alpha=0.3)

z_female = np.polyfit(female_data['total_bill'], female_data['tip'], 1)
p_female = np.poly1d(z_female)
ax2.plot(female_data['total_bill'], p_female(female_data['total_bill']), "k--", 
         linewidth=2, alpha=0.8, label='Линия регрессии')

corr_female = female_data['total_bill'].corr(female_data['tip'])
ax2.text(0.05, 0.95, f'Корреляция: {corr_female:.3f}\nn = {len(female_data)}', 
         transform=ax2.transAxes, 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8),
         verticalalignment='top')

x_min = tips['total_bill'].min() - 2
x_max = tips['total_bill'].max() + 2
y_min = tips['tip'].min() - 1
y_max = tips['tip'].max() + 1

ax1.set_xlim(x_min, x_max)
ax1.set_ylim(y_min, y_max)
ax2.set_xlim(x_min, x_max)
ax2.set_ylim(y_min, y_max)

plt.tight_layout()
plt.show()
```
